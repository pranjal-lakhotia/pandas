### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user','|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def isMale(x):
    if x == 'M':
       return 1
    return 0
users['gender_count'] = users.gender.apply(isMale)
a = users.groupby('occupation').gender_count.sum()/users.occupation.value_counts()*100
a.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by = ['occupation','gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_map_to_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
total_professionals_per_occupation =  users.groupby(['occupation']).agg('count')
# total_professionals_per_occupation.dtype
ans = gender_map_to_ocup.div(total_professionals_per_occupation,level = 'occupation')*100
ans['gender']
```
